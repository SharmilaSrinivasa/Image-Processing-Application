```python
#read image
img_f = cv2.imread('./face.jpg')
#convert to gray scale
gray_s = cv2.cvtColor(img_f, cv2.COLOR_BGR2GRAY)
#apply haar cascade
haar = cv2.CascadeClassifier('./haarcascade_frontalface_default.xml')
faces_s = haar.detectMultiScale(gray_s,1.3,5)
```

```python
print(faces_s)
```

```python
cv2.rectangle(img_f,(160,84),(160+179,84+179),(0,255,0),5)
plt.imshow(img_f)
```

```python
cv2.imshow('object_detect', img_f)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
#crop
face_crop = img_f[84:84+179,160:160+179]
plt.imshow(face_crop)
```

```python
cv2.imwrite('./female.png',face_crop)
```
