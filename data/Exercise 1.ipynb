```python
r = img_mat[:,:,0]
g = img_mat[:,:,1]
b = img_mat[:,:,2]
r.shape, g.shape, b.shape
```

```python
plt.figure(figsize = (10,6))
plt.subplot(1,3,1)
plt.title('Red region')
plt.imshow(r, cmap='gray')
plt.subplot(1,3,2)
plt.title('Green region')
plt.imshow(g, cmap='gray')
plt.subplot(1,3,3)
plt.title('Blue region')
plt.imshow(b, cmap='gray')
plt.show()
```
